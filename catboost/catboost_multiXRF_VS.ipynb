## Our py files

```python
from catboostmodels import catboost_multi_xrf
from defaults import framecleaner, splitterz
```

# dataframes

```python
df0 = pd.read_csv('../../core_to_wl_merge/OS0_Merged_dataset_imputed_08_23_2021.csv')

df1 = pd.read_csv('../../core_to_wl_merge/OS1_Merged_dataset_imputed_08_23_2021.csv')

df2 = pd.read_csv('../../core_to_wl_merge/OS2_Merged_dataset_imputed_08_23_2021.csv')
```

```python
param_dict ={
    "iter" : 1000,       #Maximum iterations, should be 500-1500
    "GPUorCPU": "CPU", #CPU is fine for this dataset on most laptops
    "dataset":['CAL', 'GR', 'DT', 'SP', 'DENS', 'PE', 'RESD', 'PHID', 'Ti', 'Mg', 'Al', 'Ca', 'Si'],
    "inputs": ['CAL', 'GR', 'DT', 'SP', 'DENS', 'PE', 'RESD', 'PHID'],
    "target": ['Ti', 'Mg', 'Al', 'Ca', 'Si' ]}
```

# Offset 0

```python
X, Y_array = framecleaner(
                            df0, 
                            param_dict['dataset'],  
                            param_dict['inputs'], 
                            param_dict['target'] )

X_train, X_test, y_train, y_test = splitterz(X.values, Y_array)

df_OS0 = catboost_multi_xrf(X_train, X_test, y_train, y_test, param_dict['iter'], OS='OS0')
```

```python
df_OS0
```

# Offset 1

```python
X1, Y1_array = framecleaner(
                            df1, 
                            param_dict['dataset'],  
                            param_dict['inputs'], 
                            param_dict['target'] )

X_train1, X_test1, y_train1, y_test1 = splitterz(X1.values, Y1_array)

df_OS1 = catboost_multi_xrf(X_train1, X_test1, y_train1, y_test1, param_dict['iter'], OS='OS1')
```

```python
df_OS1
```

# Offset 2

```python
X2, Y2_array = framecleaner(
                            df2, 
                            param_dict['dataset'],  
                            param_dict['inputs'], 
                            param_dict['target'] )

X_train2, X_test2, y_train2, y_test2 = splitterz(X2.values, Y2_array)

df_OS2 = catboost_multi_xrf(X_train2, X_test2, y_train2, y_test2, param_dict['iter'], OS='OS2')
```

```python
df_OS2
```

# Output and export

```python
frames = [df_OS0, df_OS1, df_OS2]
results = pd.concat(frames)
results
```
